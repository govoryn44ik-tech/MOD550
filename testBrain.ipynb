```python
import mne
import pandas as pd

# Read the EDF file into an MNE Raw object
raw = mne.io.read_raw_edf("1 (1).edf", preload=True)

# Get the data and channel names from the Raw object
data = raw.get_data()
channel_names = raw.ch_names

# Create a pandas DataFrame
df = pd.DataFrame(data.T, columns=channel_names)

# Display the DataFrame
print(df.head())
```

```python
import matplotlib.pyplot as plt
df.plot(x='TIME_STAMP_s', y='AF3', figsize=(15,6))
plt.title('A3_Solving')
plt.xlabel('time.sec')
plt.ylabel('A3')
plt.show
```

```python
import matplotlib.pyplot as plt
df.plot(x='TIME_STAMP_s', y='T7', figsize=(15,6))
plt.title('Signal dependence on time for T7')
plt.xlabel('time.sec')
plt.ylabel('T7')
plt.show
```
