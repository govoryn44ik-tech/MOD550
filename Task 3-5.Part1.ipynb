```python
def line_plot(self, x, y, xlabel="X", ylabel="Y"):
        plt.figure()
        plt.plot(x, y, marker='o')
        plt.title(self.title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.grid(True)
        plt.show()
```
