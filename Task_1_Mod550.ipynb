Task 1

Create a histogram from a 2d random distribution

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.randn(10000)
y = np.random.randn(10000)
plt.hist2d(x,y, bins=30)
plt.colorbar(label='Number of points')
plt.title('2D Histogram of Random Data')
plt.show
```

Task 2

Make a 2d heat map from a 2d random distribution

```python
import numpy as np
import matplotlib.pyplot as plt

#Step 1: Generate a 2D random distribution.
# We create two arrays of 10,000 random numbers each from a normal distribution.
np.random.seed(42) # This ensures the same random numbers are generated every time.
x = np.random.randn(10000)
y = np.random.randn(10000)

# Step 2: Create a 2D heat map using hist2d.
# The hist2d function calculates the frequency of data points in a 2D grid.
fig, ax = plt.subplots(figsize=(8, 6))
heatmap, xedges, yedges = np.histogram2d(x, y, bins=50)
im = ax.imshow(
    heatmap.T,
    interpolation='nearest',
    origin='lower',
    extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]],
    cmap='plasma'
)

# Step 3: Add a color bar and labels.
# The color bar shows what the colors on the map represent.
fig.colorbar(im, ax=ax, label='Frequency')
ax.set_title('2D Heat Map of a Random Normal Distribution')
ax.set_xlabel('X-axis values')
ax.set_ylabel('Y-axis values')

# Step 4: Save the plot to a file.
plt.savefig('2d_heatmap.png')
print("2D heat map has been created and saved as 2d_heatmap.png.")
```
